## in Plotly

```python
import plotly.express as px
```

```python
import plotly.express as px

fig = px.histogram(
    df_hotels, 
    x="price",
    nbins=30,
    title="Hotel Price Distribution (USD)",
    labels={"price": "Price (USD)"}
)
fig.show()
```

```python
# The price distribution of Amsterdam hotels is heavily concentrated between $200 and $500, indicating a strong mid-range market. 
# A small number of premium hotels charge above $600 and pushing toward $1800. 
# Budget hotels below $200 and ultra-luxury listings above $1200 are comparatively rare. 
# Overall, the market shows a dense middle segment with only a limited supply of low-cost or high-end options.
```

```python
fig = px.histogram(
    df_hotels,
    x="review_score",
    nbins=10,
    title="Review Score Distribution",
    labels={"review_score": "Review Score"}
)
fig.show()
```

```python
# The distribution of review scores shows that most Amsterdam hotels achieve strong guest satisfaction, with the majority scoring between 7.5 and 9.0. 
# Low-rated properties below 7.0 are rare, indicating generally high accommodation standards across the city. 
# A small group of hotels reaches exceptional scores above 9.0, while overall the distribution leans toward higher ratings.
```

```python
fig = px.bar(
    df_hotels["stars"].value_counts().sort_index(),
    title="Star Rating Distribution",
    labels={"index": "Stars", "value": "Number of Hotels"}
)
fig.show()
```

```python
# The star rating distribution shows that Amsterdam’s hotel market is dominated by mid-range and upper-mid-range properties, with 3-star and 4-star hotels making up the majority of listings. 
# Luxury 5-star hotels represent a smaller but significant segment, while budget options such as 1- and 2-star hotels are relatively small.
# The presence of many unrated properties also highlights the large number of apartments, hostels, and boutique-style accommodations in the city.
```

```python
fig = px.scatter(
    df_hotels,
    x="stars",
    y="price",
    title="Price Distribution by Star Rating",
    labels={"stars": "Stars", "price": "Price (USD)"}
)
fig.show()
```

```python
fig = px.scatter(
    df_hotels,
    x="longitude",
    y="latitude",
    color="price",
    size="price",
    title="Hotel Locations in Amsterdam Colored by Price",
    labels={"longitude": "Longitude", "latitude": "Latitude", "price": "Price (USD)"},
    hover_name="name",
    color_continuous_scale="thermal"
)
fig.update_yaxes(scaleanchor="x", scaleratio=1)  # keep aspect ratio
fig.show()
```

```python
fig = px.scatter_mapbox(
    df_hotels,
    lat="latitude",
    lon="longitude",
    color="price",
    size="price",
    hover_name="name",
    hover_data={"price": True, "stars": True, "review_score": True},
    color_continuous_scale="thermal",
    size_max=20,
    zoom=11.5,  # good zoom level for Amsterdam
    title="Hotel Locations in Amsterdam Colored by Price",
)

fig.update_layout(
    mapbox_style="open-street-map", 
        height=1000,
        width=1000
)

fig.show()
```

```python
# SABINA COMMENT: When using gradients on heatmaps, please use divergent gradients 
# (i.e. gradients that start in one colour and end in the opposite colour)
# only when you have both NEGATIVE & POSITIVE values
# When you have ONLY NEGATIVE or ONLY POSITIVE values, please stick to monochrome gradients
# This way, I can easily understand: darker shade = higher price
# Here, I see blue and yellow and get confused, which is why
# I'm glad you also used size as a way to indicate price, it's
# much less confusing
```

```python
# The highest-priced hotels (yellow/orange markers) are mainly concentrated in Amsterdam’s historic city center, where demand and tourism activity are strongest.
# Mid-priced hotels (purple/red markers) form a wide ring around the city center, showing a gradual drop in price as you move outward.
# The lowest-priced hotels (dark blue markers) are mostly located outside the central canal belt.
# This pattern reflects a classic location premium: the more central and scenic the location, the higher the price.
# A few expensive outliers appear outside the center, likely boutique or luxury properties positioned in quieter upscale neighborhoods.
# A few cheaper hotels are also located in the city center, typically because they are hostels, older budget properties, have fewer amenities, or compete aggressively in a dense market to attract price-sensitive travelers.
# Overall, the map shows that location strongly influences price, with the center commanding a premium—but exceptions exist due to hotel type, quality, and market strategy.
```

```python
fig = px.scatter_mapbox(
    df_hotels,
    lat="latitude",
    lon="longitude",
    color="stars",              # color by star rating
    size="stars",               # size by star rating
    hover_name="name",
    hover_data={"stars": True, "price": True, "review_score": True},
    color_continuous_scale="magma",   # good scale for categorical-ish values
    size_max=15,
    zoom=11.5,
    title="Hotel Locations in Amsterdam Colored by Star Rating",
)

fig.update_layout(
    mapbox_style="open-street-map",
        height=1000,
        width=1000
)

fig.show()
```

```python
# 4-star hotels form the largest group, and they are distributed evenly across the entire city, showing that this category is widespread and not limited to specific neighbourhoods.
# 3-star hotels are the second largest group, also spread broadly, but they appear more concentrated in the city center, and toward the south and west, where many mid-range tourist accommodations are located.
# 5-star hotels cluster mainly around the historic city center, close to major attractions, canals, and premium shopping streets.
# Lower-star hotels (0–2 stars) also tend to cluster near the historic center, likely due to the high demand in that area even for budget options.
# Across all categories, the strongest overall clustering of hotels is around the historic canal belt and the Museum Quarter, which are the key tourism zones of Amsterdam.
```

```python
avg_price_by_stars = (
    df_hotels.groupby("stars", as_index=False)["price"].mean()
)

fig = px.bar(
    avg_price_by_stars,
    x="stars",
    y="price",
    title="Average Price by Star Rating",
    labels={"stars": "Stars", "price": "Average Price (USD)"}
)
fig.show()
```
