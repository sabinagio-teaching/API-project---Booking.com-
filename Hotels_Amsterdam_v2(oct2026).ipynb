```python
url = "https://booking-com15.p.rapidapi.com/api/v1/hotels/searchHotels"

params = {
    "dest_id": -2140479,       # Amsterdam city
    "search_type": "city",
    "arrival_date": "2026-10-23",   # date in the future to make sure to select all avaliable hotels
    "departure_date": "2026-10-24",
    "adults": 2,
    "room_qty": 1,
    "page_number": 1
}

response = requests.get(url, headers=headers, params=params)
response.raise_for_status() #to raise a an error if there is a server problem
data = response.json()

print(data)

```

```python
import json
print(json.dumps(data, indent=4))
#this makes the data readable and structured view
```

```python
first_hotel = data["data"]["hotels"][1]   # second hotel from the list for an overview
print(json.dumps(first_hotel, indent=4))
```

```python
# the list of hotels from the JSON
hotels_20 = data["data"]["hotels"]

# How many hotels are on this page?
print("Number of hotels on this page:", len(hotels_20))

# check the first hotel object
print(json.dumps(hotels_20[0], indent=4))
```

```python
rows = []

for h in hotels_20:
    p = h.get("property", {})
    
    # id
    hotel_id = h.get("hotel_id")
    
    # name
    name = p.get("name")

    # ranking position(booking.com ranking system)
    ranking = p.get("rankingPosition")
    
    # stars / class
    stars = p.get("accuratePropertyClass")
    if stars is None:
        stars = p.get("propertyClass")
    
    # reviews
    review_score = p.get("reviewScore")
    review_count = p.get("reviewCount")
    
    # location
    latitude = p.get("latitude")
    longitude = p.get("longitude")
    
    # price
    price_info = p.get("priceBreakdown", {})
    gross = price_info.get("grossPrice", {})
    price = gross.get("value")
    price_currency = gross.get("currency")
    
    rows.append({
        "hotel_id": hotel_id,
        "name": name,
        "ranking_position": ranking,
        "stars": stars,
        "review_score": review_score,
        "review_count": review_count,
        "price": price,
        "currency": price_currency,
        "latitude": latitude,
        "longitude": longitude
    })

df_hotels_20 = pd.DataFrame(rows)
df_hotels_20.head(20)
```
